```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as stats
```

## Test der Solarpotenziale der Kantone Zürich, Bern und Jura

Es soll vergelichen werden, ob die Solarpotenziale in den verschiednen Kantonen unterschiedet. die 0-Hyptohese besagt, dass kein Unterschied vorliegt. Gemäss der Alternativhypothese liegt ein Unterschied zwischen den Solarpotenzialen vor. 

```python
dfsolarpotenziale = pd.read_csv('../Quelldateien/solarpotential.csv')
dfsolarpotenziale
```

```python
dfsolarsubsets = dfsolarpotenziale[['Canton', 'Scenario1_RoofsOnly_PotentialSolarElectricity_GWh' ]]
options = ['Zürich', 'Bern', 'Jura']
dfsolarsubsets = dfsolarsubsets.loc[dfsolarsubsets['Canton'].isin(options)]

dfsolarsubsets
```

Dieser Graph beschreibt die verschiednen Verteilungen der Solarpotenziale dar. Aufgrund der Starken ausreisser, werden die Boxplots stark verkleinert

```python
# Plot boxplot with groups
plt.figure(figsize=(8,2))
ax = sns.boxplot(x="Scenario1_RoofsOnly_PotentialSolarElectricity_GWh",
                 y="Canton",
                 data=dfsolarsubsets,
                 orient='h')
plt.grid()
plt.show()
```

```python
dfsolarsubsets1 = dfsolarsubsets.loc[dfsolarsubsets['Scenario1_RoofsOnly_PotentialSolarElectricity_GWh'] < 100]
dfsolarsubsets1
```

```python
# Plot boxplot with groups
plt.figure(figsize=(8,2))
ax = sns.boxplot(x="Scenario1_RoofsOnly_PotentialSolarElectricity_GWh",
                 y="Canton",
                 data=dfsolarsubsets1,
                 orient='h')
plt.grid()
plt.show()
```

```python
# Create subsets (groups)
Zürich = dfsolarsubsets.loc[dfsolarsubsets['Canton'] == 'Zürich']
Bern = dfsolarsubsets.loc[dfsolarsubsets['Canton'] == 'Bern']
Jura = dfsolarsubsets.loc[dfsolarsubsets['Canton'] == 'Jura']

# Create ANOVA
fvalue, pvalue = stats.f_oneway(Zürich['Scenario1_RoofsOnly_PotentialSolarElectricity_GWh'],
                                Bern['Scenario1_RoofsOnly_PotentialSolarElectricity_GWh'], 
                                Jura['Scenario1_RoofsOnly_PotentialSolarElectricity_GWh'])

# Print results
print('F-value:', fvalue.round(3), 'p-value', pvalue.round(4))
```

Da sich der p-Wert unter 5% befindet, kann die 0 Hypothese verworfen werden. Es bestehen also Unterschiede zwischen dem Durchschnittlichen Solarpotenzialen in den beobachteten Kantonen.
