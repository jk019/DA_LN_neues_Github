# Imports

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

# Import Data

```python
Solar_potential = pd.read_csv('http://www.uvek-gis.admin.ch/BFE/ogd/52/Solarenergiepotenziale_Gemeinden_Daecher_und_Fassaden.csv') # Datenbezug von uvek-gis.admin.ch
Solar_potential.head(5)
```

```python
print(Solar_potential.shape)
```

```python
plz = pd.read_csv('../Quelldateien/plz_gemeinde.csv') #File with all PLZ and the corresponding "Gemeinde"
plz.head(5)
```

```python
print(plz.shape)
```

# Edit the data

```python
plz['PLZ4'].replace('', np.nan, inplace=True)
plz['PLZ4'].replace('0', np.nan, inplace=True)
plz.dropna(subset=['PLZ4'], inplace=True)
plz = plz.rename({'GDENAMK': 'MunicipalityName'}, axis=1, inplace=False)
```

```python
plz['PLZ4']=plz['PLZ4'].astype(int)
plz.head(4)
```

```python
df_merged = pd.merge(left=Solar_potential, right=plz, how='left', left_on='MunicipalityName', right_on='MunicipalityName')
df_merged.head(5)
```

```python
print(df_merged.shape)
```

```python
df = df_merged.drop_duplicates(subset=['MunicipalityName'])
df.head(3)
```

```python
print(df.shape)
```

## exlude values from Liechtenstein

```python
index = df[ (df['Country'] == 'LI') ].index
df.drop(index , inplace=True)
```

```python
df
```

# Safe DF as a CSV

```python
df.to_csv('../Ausgabedateien/solarpotential.csv')
```
