# Imports

```python
import pandas as pd
import numpy as np
```

# read the DFs

```python
autos = pd.read_csv('../Quelldateien/autos_scraped.csv') #csv of scraped EV-listings
autos.head(5)
print(autos.shape)
```

```python
plz = pd.read_csv('../Quelldateien/plz_gemeinde.csv') #csv with Names of "Gemeinden" and Postalcodes
print(plz.shape)
plz.head(5)
```

# Prepare the DFs

```python
print(plz.shape)
plz = plz.drop_duplicates('PLZ4')
plz
```

```python
df_merged_one = pd.merge(left=autos, right=plz, how='left', left_on='plz', right_on='PLZ4')
print(df_merged_one.shape)
df_merged_one
```

```python
BFS_Nummer = pd.read_excel('../Quelldateien/Gemeindestand.xlsx') #file with codes of all Bezirke
print(BFS_Nummer.shape)
BFS_Nummer.head(5)
```

```python
df_merged = pd.merge(left=df_merged_one, right=BFS_Nummer, how='left', left_on='GDENAMK', right_on='Gemeindename')
print(df_merged.shape)
df_merged.head(5)
```

```python
print(df_merged.isna().sum())
df_merged = df_merged.dropna(subset=['PLZ4', 'BFS Gde-nummer'])
```

```python
df_merged = df_merged.astype({"BFS Gde-nummer":int})
```

```python
df_merged.head(2)
```

# Safe DF as a new csv

```python
df_merged.to_csv('../Ausgabedateien/autos_aufbereitet.csv')
```
