# Import JSON of Charging points
<a id='section_2'></a>

```python
# Get current working directory
print('Current working directory:', os.getcwd())

#read data and save as dataframe
raw_data = json.load(open('../Quelldateien/Ladestellen.json')) #Json from opendata.swiss
df_raw_alpha = pd.DataFrame(raw_data["EVSEData"])
df_raw_alpha

```

# Dataframes für jeden Operator erstellen
<a id='section_3'></a>

```python
oper_list = df_raw_alpha["OperatorName"].values.tolist()

dataframes = []
for i in range(len(oper_list)):
     locals()['df' + str(i) ] = pd.DataFrame(df_raw_alpha.loc[i, 'EVSEDataRecord'])
     locals()['df' + str(i) ]['OperatorName'] = oper_list[i-1]
     dataframes.append(locals()['df' + str(i) ])
```

# Alle Daten in ein Dataframe speichern
<a id='section_4'></a>

```python
df_raw = pd.concat(dataframes)
df_raw
```

# Koordinaten auslesen
<a id='section_5'></a>

##### hier versuchen eigenen Regex zu bauen

```python
# Getting single values from as string
coordinates = df_raw.GeoCoordinates

# Now loop through the coordinates object
out_list = []
for i in coordinates:
    # Getting single values from as string
    coord_str = "".join(i.values())
    # Splitting string by white space
    coords = re.findall(r'\S+', coord_str)
    out_list.append(coords)
    
out_list

# Create data frame from the list
df = pd.DataFrame(out_list, columns =['Latitude', 'Longitude'])
print(df.dtypes)
print(df)

# Here we have still strings of the which need to be converted to integers

# First: exclude white space using pythons lambda function
df['Latitude'] = df['Latitude'].apply(lambda x: x.strip())
df['Longitude'] = df['Longitude'].apply(lambda x: x.strip())

# Second: cange to integer values
df['Latitude'] = df['Latitude'].astype(float)
df['Longitude'] =  df['Longitude'].astype(float)
print(df.dtypes)
```
